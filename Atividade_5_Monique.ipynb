```python
def listar_alunas(alunas):

    if not alunas:

        return "Não há registros."

    return "\n".join([f"{aluna.nome} {aluna.sobrenome}" for aluna in alunas])



def consultar_faltas(alunas, nome_completo):

    for aluna in alunas:

        if f"{aluna.nome} {aluna.sobrenome}" == nome_completo:

            faltas = aluna.presenca.count(False)

            return f"Quantidade de faltas: {faltas}"

    return "Aluna não encontrada."



def menu():

    alunas = []

    while True:

        print("\nMenu:")

        print("1. Adicionar nova aluna")

        print("2. Listar alunas")

        print("3. Consultar faltas de uma aluna")

        print("4. Sair")

        opcao = input("Escolha uma opção: ")



        if opcao == "1":

            nome = input("Nome: ")

            sobrenome = input("Sobrenome: ")

            turma = input("Turma: ")

            try:

                notas = list(map(float, input("Notas (separadas por espaço): ").split()))

                presenca = list(map(lambda x: x.lower() == 'true', input("Presença (True/False separados por espaço): ").split()))

                participacao = float(input("Nota de participação: "))

                print(adicionar_aluna(alunas, nome, sobrenome, turma, notas, presenca, participacao))

            except ValueError:

                print("Erro: Dados inválidos. Certifique-se de que as notas e a participação são números e a presença é True ou False.")



        elif opcao == "2":

            print(listar_alunas(alunas))



        elif opcao == "3":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_faltas(alunas, nome_completo))



        elif opcao == "4":

            print("Saindo...")

            break



        else:

            print("Opção inválida. Tente novamente.")



def menu():

    alunas = []

    while True:

        print("\nMenu:")

        print("1. Adicionar nova aluna")

        print("2. Listar alunas")

        print("3. Consultar faltas de uma aluna")

        print("4. Consultar notas de uma aluna")

        print("5. Sair")

        opcao = input("Escolha uma opção: ")



        if opcao == "1":

            nome = input("Nome: ")

            sobrenome = input("Sobrenome: ")

            turma = input("Turma: ")

            try:

                notas = list(map(float, input("Notas (separadas por espaço): ").split()))

                presenca = list(map(lambda x: x.lower() == 'true', input("Presença (True/False separados por espaço): ").split()))

                participacao = float(input("Nota de participação: "))

                print(adicionar_aluna(alunas, nome, sobrenome, turma, notas, presenca, participacao))

            except ValueError:

                print("Erro: Dados inválidos. Certifique-se de que as notas e a participação são números e a presença é True ou False.")



        elif opcao == "2":

            print(listar_alunas(alunas))



        elif opcao == "3":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_faltas(alunas, nome_completo))



        elif opcao == "4":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_notas(alunas, nome_completo))



        elif opcao == "5":

            print("Saindo...")

            break



        else:

            print("Opção inválida. Tente novamente.")



def menu():

    alunas = []

    while True:

        print("\nMenu:")

        print("1. Adicionar nova aluna")

        print("2. Listar alunas")

        print("3. Consultar faltas de uma aluna")

        print("4. Consultar notas de uma aluna")

        print("5. Consultar status de aprovação de uma aluna")

        print("6. Sair")

        opcao = input("Escolha uma opção: ")



        if opcao == "1":

            nome = input("Nome: ")

            sobrenome = input("Sobrenome: ")

            turma = input("Turma: ")

            try:

                notas = list(map(float, input("Notas (separadas por espaço): ").split()))

                presenca = list(map(lambda x: x.lower() == 'true', input("Presença (True/False separados por espaço): ").split()))

                participacao = float(input("Nota de participação: "))

                print(adicionar_aluna(alunas, nome, sobrenome, turma, notas, presenca, participacao))

            except ValueError:

                print("Erro: Dados inválidos. Certifique-se de que as notas e a participação são números e a presença é True ou False.")



        elif opcao == "2":

            print(listar_alunas(alunas))



        elif opcao == "3":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_faltas(alunas, nome_completo))



        elif opcao == "4":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_notas(alunas, nome_completo))



        elif opcao == "5":

            nome_completo = input("Digite o nome completo da aluna: ")

            print(consultar_status_aprovacao(alunas, nome_completo))



        elif opcao == "6":

            print("Saindo...")

            break



        else:

            print("Opção inválida. Tente novamente.")



if __name__ == "__main__":

    menu()
```
